```python
#ordinal tanımlama
from pandas.api.types import CategoricalDtype
```

```python
df.cut.head()
```

```python
df.cut = df.cut.astype(CategoricalDtype(ordered = True))
```

```python
df.dtypes
```

```python
df.cut.head(1)
```

```python
#sıralama işlemi ters oldu kendisi uygun olduğu şekilde sıralıyor bunu doğru sıraya getirmemiz gerek
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]  #küçükten büyüğe doğru sıralıyoruz 
```

```python
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
```

```python
df.cut.head()
```

```python
#barplot
```
