```python
from sklearn import preprocessing
eksikveri = pd.read_csv("eksikveriler.csv")
```

```python
#KNN Algoritması için tüm değişkenler bir sayı olmalı bu yüzden ulkeleri numeriğe çeviridk
#bunu yaparken de 1-2-3 gibi numara vermedik eğer verseydik aralarında bir numerik üstünlük olacaktı
eksikveri = pd.get_dummies(eksikveri,columns = ["ulke"],prefix = ["ulke"]) 
eksikveri
```

```python
lbe = preprocessing.LabelEncoder()
```

```python
eksikveri["y_cinsiyet"] = lbe.fit_transform(eksikveri["cinsiyet"])
```

```python
eksikveri = eksikveri.drop('cinsiyet', axis = 1)
```

```python
eksikveri
```

```python
eksikveri_columns = list(eksikveri)
```

```python
eksikveri = np.array(eksikveri)
```

```python
dff = knnimput.KNN(k=4).complete(eksikveri)
```

```python
KNN_doldurma = pd.DataFrame(dff,columns = eksikveri_columns)
```
